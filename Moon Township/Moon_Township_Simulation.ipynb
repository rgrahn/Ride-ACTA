```python
# ============== Load in travel time information obtained from Google =================
y = np.load('googleTravTimes_REVISED.npy', allow_pickle = True)
googleTravTimes = y[()]

# Select from csv output from Demand_Sampling_Moon file based on expected demand level
fileName = '.csv'
moonRidership = pd.read_csv(fileName)
# Load in Moon Business Information
moonLocations = pd.read_csv('All_Moon_Businesses_REVISED.csv')
moonLocations1 = moonLocations[['Latitude','Longitude','Cluster']].copy()

def createDfFromODMatrices(googleTravTimes):
    df = pd.DataFrame(columns=['timePeriod', 'orig','dest','origLat','origLong','destLat','destLong','travTime'])
    for key, value in googleTravTimes.items():
        for i in range(len(value)):
            for j in range(len(value[0])):
                toAppend = [key, moonLocations['Cluster'].iloc[i], moonLocations['Cluster'].iloc[j], \
                                     moonLocations['Latitude'].iloc[i], moonLocations['Longitude'].iloc[i], \
                                     moonLocations['Latitude'].iloc[j], moonLocations['Longitude'].iloc[j], value[i, j]]
                appendSeries = pd.Series(toAppend, df.columns)
                df = df.append(appendSeries, ignore_index=True)
    return df

def createTravTimeDict(data):
    travTimes = dict()
    for timePeriod in data['timePeriod'].unique():
        filtData = data[data['timePeriod'] == timePeriod].copy()
        G = nx.from_pandas_edgelist(filtData, source = 'orig', target = 'dest', edge_attr = 'travTime', create_using = nx.DiGraph())
        travTimes[timePeriod] = G
    return travTimes

df = createDfFromODMatrices(googleTravTimes)
df1 = df.dropna()
timeDepTravTime = createTravTimeDict(df1)
```

```python
# ====== Format data for simulation =================
# Join origin and destination coordinates
df1 = pd.merge(moonRidership, moonLocations1, how = 'left', left_on = 'orig', right_on = 'Cluster')
df1 = df1.rename(columns = {'Latitude':'origLat', 'Longitude':'origLong','Cluster':'origCluster'})
df2 = pd.merge(df1, moonLocations1, how = 'left', left_on = 'dest', right_on = 'Cluster')
df2 = df2.rename(columns = {'Latitude':'destLat', 'Longitude':'destLong','Cluster':'destCluster'})

# Create a time period column
def getTimePeriodAndSecs(data):
    data.loc[:, 'timePeriod'] = 0
    data.loc[:, 'origSec'] = 0
    data.loc[:, 'hour'] = 0
    for i in range(len(data)):
        splitTimes = data['deptTime'].iloc[i].split(':')
        data.loc[i, 'timePeriod'] = splitTimes[0] + ':00:00'
        minute = '00' if int(splitTimes[1])//30 == 0 else '30' 
        data.loc[i, 'timePeriod1'] = splitTimes[0] + ':' + minute + ':00'
        data.loc[i, 'origSec'] = int(splitTimes[0])*3600 + int(splitTimes[1])*60 + int(splitTimes[2])
        data.loc[i, 'hour'] = int(splitTimes[0])
    return data

riderData = getTimePeriodAndSecs(df2) 
newDF = riderData[['ride_id', 'origSec']].copy()
origTimeDict = newDF.set_index('ride_id').to_dict()
riderTimeDict = origTimeDict['origSec']
```

```python
# =============== Multi-day Simulation ===============

# import MOON_RouteSimClass as simul
# from MOON_RouteSimClass import *
# import MOON_RouteRiderVanClass as rv
# from MOON_RouteRiderVanClass import *
# importlib.reload(simul)
# importlib.reload(rv)

import MOON_RouteSimClassFixed as simul1
from MOON_RouteSimClassFixed import *
import MOON_RouteRiderVanClassFixed as rv1
from MOON_RouteRiderVanClassFixed import *
importlib.reload(simul1)
importlib.reload(rv1)


# Create a list of rider objects to serve during the day
def createDailyRiderList(riderData):
    riderList = []
    for i in range(len(riderData)):
        riderList.append(rider(riderData['ride_id'].iloc[i], riderData['origSec'].iloc[i], 
                               riderData['orig'].iloc[i], riderData['dest'].iloc[i], 
                               riderData['timePeriod'].iloc[i],
                               riderData['origLat'].iloc[i], riderData['origLong'].iloc[i],
                               riderData['destLat'].iloc[i], riderData['destLong'].iloc[i]))
    # create a list of riders that need to be served where riders can be removed throughout the day
    remainingRiders = riderList.copy()
    return riderList, remainingRiders

def storeSimulationResults(dailyRiders, riderData, sample, timeDepTravTime):
    # Filter original rider data
    dfVan = riderData[(riderData['sample'] == sample)].copy()
    # Convert new dataframe from the simulation results
    a,b,c,d =[],[],[],[]
    for rider in dailyRiders:
        a.append(rider.van)
        b.append(rider.pickupTime)
        c.append(rider.dropoffTime)
        d.append(rider.rideID)
    x = pd.DataFrame(data = {'ride_id':d,'vanID':a, 'pickup_Sim':b,'dropoff_Sim':c})
    mergeDF = pd.merge(dfVan, x, how = 'left', on = 'ride_id')
    mergeDF.loc[:,'waitTimeSim'] = mergeDF['pickup_Sim'] - mergeDF['origSec']
    mergeDF.loc[:,'driveTimeSim'] = mergeDF['dropoff_Sim'] - mergeDF['pickup_Sim']
    # For Uber Case
    for i in range(len(mergeDF)):
        if (pd.isnull(mergeDF['pickup_Sim'].iloc[i])):
            mergeDF['pickup_Sim'].iloc[i] = mergeDF['origSec'].iloc[i] + 5*60
            mergeDF['waitTimeSim'].iloc[i] = 5*60
            travTime = nx.shortest_path_length(timeDepTravTime[mergeDF['timePeriod'].iloc[i]], 
                                    source = mergeDF['orig'].iloc[i], 
                                    target = mergeDF['dest'].iloc[i], weight = 'travTime')
            mergeDF['driveTimeSim'].iloc[i] = travTime
    return mergeDF

def processSimulationData(simResults, sample):
    simResults.loc[(simResults['waitTimeSim'] < 0), 'waitTimeSim'] = 0
    simResults['waitTimeSim'] = simResults['waitTimeSim'].astype(float)
    simResults['driveTimeSim'] = simResults['driveTimeSim'].astype(float)
    simResults = simResults.rename(columns={'waitTimeSim':'simulWaitTime', 'driveTimeSim':'simulDriveTime'})
    return simResults

def runDaySimulation(sample, timeDepTravTime, riderTimeDict, vehCount, timeWindow, uberWait, aggregate=False):
    filtData = riderData[riderData['sample'] == sample].reset_index(drop=True).copy()
    if (aggregate == True):
        filtData[['orig','dest']] = filtData[['orig','dest']].replace(26, 22)
        filtData[['orig','dest']] = filtData[['orig','dest']].replace(16, 17)
        filtData[['orig','dest']] = filtData[['orig','dest']].replace(23, 17)
        filtData[['orig','dest']] = filtData[['orig','dest']].replace(24, 17)
        #filtData[['orig','dest']] = filtData[['orig','dest']].replace(13, 21)
        filtData[['orig','dest']] = filtData[['orig','dest']].replace(19, 20)
        
    dailyRiders, remainingRiders = createDailyRiderList(filtData)

    # ---------------------
    simStartTime = 3600*6 # <-- Initiate at 6am
    ikeaVehs = dict()
    ikeaVehs[1533] = simStartTime
    ikeaVehs[1535] = simStartTime
    if (vehCount in [1, 99]): # 99 is case for one van and Uber
        del ikeaVehs[1535]
    waitCoeff = 0.75
    driveCoeff = 0.25
    #timeWindow = 0*60

    # Inputs: {simStartTime = time to start simulation, default=6am,
    #          ikeaVehs=ikea veh numbers with start times {1533:6am (in seconds)},
    #          remainingRiders=list of ride requests for the given day,
    #          timeDepTravTime=graph of travel times between nodes,
    #          rideTimeDict= {requestID:origTime (sec)},
    #          waitCoeff, driveCoeff = how to weight wait/drive times to pick best route,
    #          timeWindow=look ahead time window}

    # ========= Initialize simulation =============
    sim = Sim(simStartTime, ikeaVehs, remainingRiders, timeDepTravTime, riderTimeDict, waitCoeff, driveCoeff, timeWindow, uberWait) 

    while (sim.time <= 19*3600-60):
        # Assign working vans based on their start times
        sim.activateVans()
        # Assign rider to van with lowest 'marginal cost' (coeff*waitTime + coeff*driveTime)
        sim.assignRiderToVan()
        sim.nextTrip()
        sim.moveVans()
        sim.step()
        # -------------- Uncomment for fixed route simulation -------------------
        #sim.assignFixedRoute()
    
    simResults = storeSimulationResults(dailyRiders, filtData, sample, timeDepTravTime)
    plotSimData = processSimulationData(simResults, sample)
    return plotSimData

def runMultipleDays(samples, timeDepTravTime, riderTimeDict, vehCount, timeWindow, uberWait, aggregate=False):
    count = 0
    for sample in samples:
        if (sample % 10 == 0):
            print(sample)
        if (count == 0):
            simulationResults = runDaySimulation(sample, timeDepTravTime, riderTimeDict, vehCount, timeWindow, uberWait, aggregate)
            count += 1
        else:
            simulationResults1 = runDaySimulation(sample, timeDepTravTime, riderTimeDict, vehCount, timeWindow, uberWait, aggregate)
            simulationResults = simulationResults.append([simulationResults1],sort=True)
    return simulationResults
```

```python
#  ==================== For Uber Only Case - calculate drive time based on time dep travel times ===================
test = a[['sample','ride_id','orig','dest','deptTime','timePeriod','vehCount','hour','vanID',\
          'origSec', 'pickup_Sim','dropoff_Sim']].copy()
uberOnly = test[test['vehCount'] == 2].copy()
uberOnly.loc[:,'vehCount'] = 3
uberOnly.loc[:,'simulWaitTime'] = 5*60
uberOnly.loc[:,'pickup_Sim'] = uberOnly['origSec'] + 5*60
uberOnly.loc[:,'simulDriveTime'] = 0
uberOnly.loc[:,'vanID'] = 'Uber'

for i in range(len(uberOnly)):
    uberOnly['simulDriveTime'].iloc[i] = nx.shortest_path_length(timeDepTravTime[uberOnly['timePeriod'].iloc[i]], \
                                                             source = int(uberOnly['orig'].iloc[i]), \
                                                             target = int(uberOnly['dest'].iloc[i]), \
                                                             weight = 'travTime')


# ============== Add Uber only results to the rest of the simulation results ===================
filtResults = a[['sample','ride_id','orig','dest','deptTime','timePeriod','vehCount',\
                 'simulWaitTime','simulDriveTime','hour','vanID', 'pickup_Sim','dropoff_Sim','origSec']].copy()
allResults = filtResults.append([uberOnly], sort=True)
allResults['ones'] = 1
allResults.loc[:,'userCost'] = allResults['simulWaitTime']*0.75 + allResults['simulDriveTime']*0.25

# Load in distance matrix
travDist = pd.read_csv('Moon_Trav_Dist.csv')

# ==================   For calculating the cost of each Uber trip ====================
def distDict(data):
    storeDict = dict()
    for i in range(len(data)):
        orig = data['Orig_Node'].iloc[i]
        dest = data['Dest_Node'].iloc[i]
        storeDict[f'{orig}, {dest}'] = data['dist'].iloc[i]
        storeDict[f'{dest}, {orig}'] = data['dist'].iloc[i]
    return storeDict

def calcCost(timeMin, distMile):
    minFare = 8
    baseFare = 1.43
    bookingFee = 2.5
    cost = baseFare + bookingFee + (0.32*timeMin) + (0.87*distMile)
    totalFee = max(minFare, cost)
    return totalFee
    
def addUberCosts(data, dist):
    data.loc[:,'uberCost'] = 0
    for i in range(len(data)):
        if (data['vanID'].iloc[i] == 'Uber'):
            timeMin = data['simulDriveTime'].iloc[i]/60
            od_pair = str(data['orig'].iloc[i]) + ', ' + str(data['dest'].iloc[i])
            distMile = dist.get(od_pair, 10)
            data['uberCost'].iloc[i] = calcCost(timeMin, distMile)
    return data
    
dist = distDict(travDist)
df_costs = addUberCosts(allResults, dist)

df_costs.to_csv('USE_MoonResults_45.csv')
```
